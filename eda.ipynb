```python
country_codes = pd.read_excel('datasets/Country-Code.xlsx')
```

```python
country_codes
```

```python
zomato_data = pd.merge(zomato_data,country_codes,on='Country Code',how='left')
```

```python
zomato_data
```

```python
countries = zomato_data['Country'].unique()
```

```python
# number of restaurants zomato has services in each country
restaurant_count_per_country = zomato_data.groupby('Country').agg({'Restaurant ID':'count'}).sort_values('Restaurant ID',ascending=False).reset_index()
restaurant_count_per_country
```

```python
'''Restaurants in Each country'''

fig,ax = plt.subplots(figsize=(20,10))

ax.bar(restaurant_count_per_country['Country'],restaurant_count_per_country['Restaurant ID'],mouseover=True)
ax.set_title("Restaurants per Country")
ax.set_xlabel("Country")
ax.set_ylabel("Restaurant Count")
plt.show()
```

```python
label_fmt = '%1.2f%%'
explode = (0.1,0,0)
plt.pie(restaurant_count_per_country['Restaurant ID'].iloc[:3],labels=restaurant_count_per_country['Country'].iloc[:3],
        radius=2,autopct=label_fmt,explode=explode)
plt.title('Restaurants in Top 3 Countries')
plt.show()
```

```python
all_rating_metrics = zomato_data[['Aggregate rating','Rating color','Rating text']].sort_values('Aggregate rating')
```

```python
''' Nummber of restaurants falling in different Rating range'''

zomato_data.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Restaurant count'})
```

```python
# Global level view of Rating text and Aggregate Rating
bins = {}

for rating in all_rating_metrics['Rating text'].unique():
    temp = all_rating_metrics[all_rating_metrics['Rating text'] == rating]
    bins[rating]=f"{temp['Aggregate rating'].min()}-{temp['Aggregate rating'].max()}"
```

```python
bins = pd.DataFrame.from_dict(bins,orient='index',columns=['rating_range'])
```

```python
'''Rating bins classified in their respective band'''

bins
```

```python
colors=['red','green','blue','cyan','magenta','yellow','black','olive','orange','purple','pink','gray','brown','olive','teal']

x_label = ['Poor','Average','Good','Very Good','Excellent','Not rated']

rating_scale = zomato_data['Rating text'].value_counts().index.to_series().map(bins['rating_range']).to_dict()
```

```python
''' Global Restaurant Count with respect to their ratings and the corresponding rating band'''

label = [rating_scale[i] for i in x_label]
fig,ax = plt.subplots(figsize=(20,10))
b_plot=ax.bar(x_label,[zomato_data['Rating text'].value_counts().to_dict().get(c,0) for c in x_label])
ax.bar_label(b_plot,label)
ax.set_title("Restaurant Ratings")
ax.set_xlabel("Rating Text")
ax.set_ylabel("Restaurant Count")
plt.show()
```

```python
''' Country level Restaurant Count with respect to their ratings and the corresponding rating band'''


for country, color in zip(zomato_data['Country'].value_counts().index,colors):
    print(color)
    data = zomato_data[zomato_data['Country']==country]
    label = [rating_scale[i] for i in x_label]
    fig,ax = plt.subplots(figsize=(10,5))
    b_plot=ax.bar(x_label,[data['Rating text'].value_counts().to_dict().get(c,0) for c in x_label],color=color)
    ax.bar_label(b_plot,label)
    ax.set_title(f"{country} Ratings")
    ax.set_xlabel("Rating Text")
    ax.set_ylabel("Restaurant Count")
    plt.show()
```

```python
''' Number of restaurants trading in different currencies '''

zomato_data.groupby('Country').Currency.value_counts().sort_values(ascending=False)
```

```python
''' Number of restaurants that have Online Delivery'''

zomato_data.groupby(['Country','Has Online delivery']).size().reset_index().sort_values('Has Online delivery').rename(columns={0:'Restaurant Count'})
```

```python
''' Top 5 cities ordered with respect to the count of Restaurants'''

city_values = zomato_data['City'].value_counts().values
city_labels = zomato_data['City'].value_counts().index

label_fmt = '%1.2f%%'
explode = (0.1,0,0,0,0)
plt.pie(city_values[:5],labels=city_labels[:5],
        radius=2,autopct=label_fmt,explode=explode)
plt.title('Top 5 Cities with Highest restaurants')
plt.show()
```
